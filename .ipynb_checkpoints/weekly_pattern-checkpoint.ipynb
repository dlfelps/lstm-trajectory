```python
import pandas as pd
from pathlib import Path
import numpy as np
import pickle
from tqdm import tqdm
```

```python
poi_file = Path('data/POIdata_cityD.csv')
df = pd.read_csv(poi_file)
df
```

```python
def distance_matrix(group):    
    locs_x = group.x.to_numpy()
    locs_y = group.y.to_numpy()
    def one_dist(x,y):
        #capture locs_x and locs_y above
        dx = locs_x - x
        dy = locs_y - y
        d = dx*dx + dy*dy
        return np.sqrt(d)
    
    D = np.zeros((200,200, len(locs_x)))
    for x in range(200):
        for y in range(200):
            D[x,y,:]=one_dist(x,y)

    return D

def distance_to_disk(folder):
    
    for c, g in tqdm(df.groupby('category')):
        temp = distance_matrix(g)
        np.savez_compressed(folder.joinpath(f"{c}.npz"), array=temp)

    return None


distance_to_disk(Path('precomputed/distance_matrix'))
```
